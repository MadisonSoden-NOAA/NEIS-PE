```python
import matplotlib.pyplot as plt
from tools import SRI_sum

mlrf1 = SRI_sum.main('mlrf1')
fwyf1 = SRI_sum.main('fwyf1')
sanf1 = SRI_sum.main('sanf1')
smkf1 = SRI_sum.main('smkf1')

print("mlrf1\n", mlrf1)
print("fwyf1\n", fwyf1)
print("sanf1\n", sanf1)
print("smkf1\n", smkf1)
```
